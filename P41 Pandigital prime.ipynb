We shall say that an n-digit number is pandigital if it makes use of all the digits 1 to n exactly once. For example, 2143 is a 4-digit pandigital and is also prime.

What is the largest n-digit pandigital prime that exists?

```python
def is_prime(n):
    root=int(n**0.5)
    
    for i in range (2,root):
        if n%i==0:
            return False
    if n%root==0:
        return False
    else:
        return True
```

```python
import itertools
def generate_pandigital(length):
    num=''
    for i in range (1, length+1):
        num+=str(i)
    
    iter_num=[''.join(p) for p in itertools.permutations(num)]
    iter_num=sorted(list(set(iter_num)))
    iter_num.reverse()
    return iter_num
```

```python
def find_highest_pandigital_num():
    for i in range (9,0, -1):
        print ('Searching length ', i) 
        pandigital_num=generate_pandigital(i)
        
        for num in pandigital_num:
            if is_prime(int(num)):
                print ('Found prime number {}'.format(num))
                return num
find_highest_pandigital_num()
```
