The Fibonacci sequence is defined by the recurrence relation:

Fn = Fn−1 + Fn−2, where F1 = 1 and F2 = 1.
Hence the first 12 terms will be:

F1 = 1
F2 = 1
F3 = 2
F4 = 3
F5 = 5
F6 = 8
F7 = 13
F8 = 21
F9 = 34
F10 = 55
F11 = 89
F12 = 144
The 12th term, F12, is the first term to contain three digits.

What is the index of the first term in the Fibonacci sequence to contain 1000 digits?



```python

def find_rounded_fibo(num_digits):
    last2_fibo=[1,1]
    num_zero=0
    idx=2
    while num_zero+10<num_digits:
        next_fibo=sum(last2_fibo)
        last2_fibo=[last2_fibo[-1], next_fibo]
        if next_fibo>1e10:
            num_zero+=1
            last2_fibo=[last2_fibo[-2]//10, last2_fibo[-1]//10]
        idx+=1
    return idx
            
            
            
```

```python
find_rounded_fibo(1000)
```

```python
def find_fibo(n):
    fibo=[1,1]
    
    while len (fibo)<n:
        fibo.append(sum(fibo[-2:]))
    return fibo
```
