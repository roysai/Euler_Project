Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

Triangle	 	Tn=n(n+1)/2	 	1, 3, 6, 10, 15, ...
Pentagonal	 	Pn=n(3n−1)/2	 	1, 5, 12, 22, 35, ...
Hexagonal	 	Hn=n(2n−1)	 	1, 6, 15, 28, 45, ...
It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.

```python
def find_triangular(limit):
    triangular=[int(n*(n+1)/2) for n in range (1, limit)]
    return triangular

def find_hexagonal(limit):
    hexagonal=[int(n*(2*n-1)) for n in range (1, limit)]
    return hexagonal

def is_pentagonal(num):
    n=(1+(24*num+1)**0.5)/6
    #print (n)
    if n%1==0:
        return True
    else:
        return False
```

```python
def find_triplet(start):
    
    while True:
        number=start*(start+1)/2
        if is_pentagonal(number):
            return int(number), start
        start+=2
```

```python
number, idx= find_triplet(287)
```
